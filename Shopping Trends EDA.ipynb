## Data Analysis

### Age Distribution

```python
df["Age"].plot(kind = "hist",bins=10, linewidth=0.3,edgecolor="white",color = "#3C6255")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.show()
```

* The chart shows a histogram representing the distribution of "Age." 
* The X-axis represents age groups, ranging from 20 to 70, with intervals of 10 years (20, 30, 40,       etc.) and the y-axis represents the frequency, or the count of occurrences for each age group. 
* Age Distribution: The age groups 20 and 70 have the highest frequency, both slightly above 400. The     frequency for the other age groups remains somewhat consistent, fluctuating between 350 to 400. There   is no significant outlier or skewness in the age distribution, indicating a relatively uniform         distribution across the age range.
* Insights: The age distribution is fairly even across different age groups, with a slight increase in   the youngest and oldest age categories (20 and 70). This could suggest a balanced representation of     age in the dataset with slightly more individuals at the extremes


### Gender Distribution

```python
plt.figure(figsize=(4, 4))
df["Gender"].value_counts().plot(kind = "bar",color = ["#16423C","#6A9C89"],width = 0.2)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.show()
```

* The chart represents the Gender Distribution using a bar plot

* The scale goes up to 2500+. 
* Gender Distribution: The male count exceeds 2500, showing a significantly larger representation in     the dataset and the female count is around 1250, which is approximately half the male count.
* Insights: The dataset has a much larger proportion of males compared to females, with males making up   roughly twice the number of females. The gender distribution is quite imbalanced.


### Subscription status

### Category wise Average review Rating

```python
## What is the average review rating for each item category?

Avg_review_rating = df.groupby("Category")["Review Rating"].mean().sort_values(ascending = False)
Avg_review_rating
```

```python
plt.figure(figsize=(6, 4))
Avg_review_rating.plot(kind = "bar",color = ["#16423C","#6A9C89","#C4DAD2","#E9EFEC"],width = 0.4)
plt.title("Category")
plt.xlabel("Review Rating")
plt.show()
```

* The bar chart above compares the average review ratings across four different product categories:       Footwear, Accessories, Outerwear, and Clothing
* Highest Average Review rating: Footwear has the highest average review rating, slightly above 3.5 and   Accessories follows closely behind with a similar average rating
* Lower Average Review rating: Outerwear is slightly lower than Accessories, but not by a significant     margin and Clothing has the lowest average rating among all of them
* The review ratings for all categories seem to hover around 3.5 out of 5. This indicates a moderate     level of satisfaction among customers. While none of the categories is rated poorly, none of them       stands out with a rating close to the maximum either
* Insights: Focusing on the Clothing category and finding ways to increase customer satisfaction could   help boost the overall average rating across all categories
* Since Footwear is leading, the company can adopt the same strategy to increase the customer             satisfaction for the rest of the categories as well

* The whiskers show that the minimum and maximum purchase amounts are around 20 and 100 for most         categories

However, since spending behavior is relatively stable across   the board, businesses could focus on encouraging more frequent purchases without worrying about a       drop in purchase amount
* The box plot indicates that purchase amounts are relatively consistent across different frequencies     of purchases. There's no significant variability in spending behavior, meaning customers tend to       spend similar amounts whether they purchase frequently (e.g., weekly) or infrequently (e.g.,           annually). ### Locations having the Highest Purchase amount

```python
## Which locations have the highest purchase amount

Location_Purchase_amount = df.groupby("Location")["Purchase Amount (USD)"].sum().head().sort_values(ascending = False)
Location_Purchase_amount
```

```python
pivot_table = df.pivot_table(values='Purchase Amount (USD)', index='Location', aggfunc='sum')

# Plotting heatmap
plt.figure(figsize=(11,11))
sns.heatmap(pivot_table, annot=True, cmap="crest")
plt.title('Location wise Total Purchase Amount')
plt.show()
```

* The heatmap above shows the total purchase amounts in USD for different U.S. states. Each state's       total purchase amount is color-coded, and the color gradient on the right side of the chart indicates   the purchase amounts from approximately 3500 USD to 5500 USD
* Top States by Purchase Amount: California (5.6e+03) and Illinois (5.6e+03) stand out with the highest   purchase amounts, indicating that customers in these states are spending the most. Montana (5.8e+03)   also has a high purchase amount, slightly higher than California and Illinois. Nevada (5.5e+03) and     Nebraska (5.2e+03) also show higher total purchase amounts.
* Lowest Purchase Amounts: States such as Kansas (3.4e+03) and Florida (3.8e+03) have some of the         lowest total purchase amounts, suggesting that customers in these states are spending less in           comparison to other regions. Hawaii (3.8e+03) and New Jersey (3.8e+03) are also on the lower end of     the spectrum.
* Mid-Range States: Most states fall within the mid-range (around 4500 to 5000 USD). States like         Minnesota (5e+03), New Mexico (5e+03), and New York (5.3e+03) are among these mid-level spenders.
* Insights: The higher spending states like Montana, California, and Illinois could indicate             stronger economic activity, higher income levels, or a higher demand for the products/services being   offered
* States with lower total purchase amounts, such as Kansas and Florida, might have smaller populations,   lower disposable incomes, or lesser demand for the products being analyzed. For states like Kansas     and Florida, understanding the underlying reasons for lower purchase amounts could inform marketing     strategies aimed at boosting sales.
* The heatmap shows a diverse distribution of total purchase amounts across different states. While       states like Montana, California, and Illinois have high purchase amounts, other regions like Kansas     and Florida report lower amounts. Businesses can use this insight to focus on high-spending regions     and devise strategies to increase sales in lower-spending states
